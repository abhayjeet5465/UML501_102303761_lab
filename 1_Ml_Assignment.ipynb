```python
# (f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find
# i. Sum of diagonal elements
# ii. Eigen values of matrix
# iii. Eigen vectors of matrix
# iv. Inverse of matrix
# v. Determinant of matrix
n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])
print("Sum of diagonal elements:", np.trace(n_array))
eigenvalues = np.linalg.eigvals(n_array)
print("Eigenvalues:", eigenvalues)
_, eigenvectors = np.linalg.eig(n_array)
print("Eigenvectors:\n", eigenvectors)
inverse = np.linalg.inv(n_array)
print("Inverse of matrix:\n", inverse)
determinant = np.linalg.det(n_array)
print("Determinant of matrix:", determinant)

```

```python
#(g) Multiply the following matrices and also find covariance between matrices using NumPy: 

p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])


product1 = np.dot(p1, q1)
print("Matrix multiplication result:\n", product1)

cov1 = np.cov(p1.T, q1.T)
print("Covariance between matrices:\n", cov1)
```

```python
p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])

q2_T = q2.T  

combined = np.hstack((p2, q2_T))  

cov_matrix = np.cov(combined.T)

print("Covariance between matrices:\n", cov_matrix)
```

Q2: Based on NumPy Mathematics and Statistics

```python
# (a) For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
# i. Find element-wise absolute value
# ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
# row.
# iii. Mean, Median and Standard Deviation of flattened array, of each column, and
# each row 




array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Absolute values
abs_array = np.abs(array)

# ii. Percentiles (Flattened)
p25_flat = np.percentile(array, 25)
p50_flat = np.percentile(array, 50)
p75_flat = np.percentile(array, 75)

# ii. Percentiles (Column-wise)
p25_col = np.percentile(array, 25, axis=0)
p50_col = np.percentile(array, 50, axis=0)
p75_col = np.percentile(array, 75, axis=0)

# ii. Percentiles (Row-wise)
p25_row = np.percentile(array, 25, axis=1)
p50_row = np.percentile(array, 50, axis=1)
p75_row = np.percentile(array, 75, axis=1)

# iii. Mean, Median, Std Dev
mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

# Output
print("Absolute values:\n", abs_array)
print("\nFlattened percentiles: 25th =", p25_flat, ", 50th =", p50_flat, ", 75th =", p75_flat)
print("Column-wise percentiles:\n25th =", p25_col, "\n50th =", p50_col, "\n75th =", p75_col)
print("Row-wise percentiles:\n25th =", p25_row, "\n50th =", p50_row, "\n75th =", p75_row)
print("\nFlattened -> Mean:", mean_flat, ", Median:", median_flat, ", Std Dev:", std_flat)
print("Column-wise -> Mean:", mean_col, ", Median:", median_col, ", Std Dev:", std_col)
print("Row-wise -> Mean:", mean_row, ", Median:", median_row, ", Std Dev:", std_row)


```

Q3: Based on Searching and Sorting

```python
#(a) For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
array = np.array([10, 52, 62, 16, 16, 54, 453])

# i. Sorted array
sorted_array = np.sort(array)

# ii. Indices of sorted array
sorted_indices = np.argsort(array)

# iii. 4 smallest elements
four_smallest = np.sort(array)[:4]

# iv. 5 largest elements
five_largest = np.sort(array)[-5:]

print("Original array:", array)
print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", four_smallest)
print("5 largest elements:", five_largest)

```

```python
#(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find 

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# i. Integer elements only (check if value is an integer using np.floor or .is_integer())
int_elements = array[np.equal(np.mod(array, 1), 0)]

# ii. Float elements only (non-integers)
float_elements = array[np.not_equal(np.mod(array, 1), 0)]

print("Original array:", array)
print("Integer elements only:", int_elements)
print("Float elements only:", float_elements)
```

```python

from PIL import Image
import os

def img_to_array(path):
    img = Image.open(path)

    if img.mode == 'L':  # Grayscale image
        img_array = np.array(img)
        np.savetxt("grayscale_image.txt", img_array, fmt='%d')
        print("Grayscale image saved as grayscale_image.txt")
    elif img.mode == 'RGB':  # RGB image
        img_array = np.array(img)
        # Save each channel separately
        for i, channel in enumerate(['R', 'G', 'B']):
            np.savetxt(f"rgb_{channel}.txt", img_array[:, :, i], fmt='%d')
        print("RGB image channels saved as rgb_R.txt, rgb_G.txt, rgb_B.txt")
    else:
        print("Unsupported image mode:", img.mode)

    return img_array

image_path = r"C:\Users\abhay\OneDrive\Desktop\JN\1_ML_LAB_102303761\stockvault-autumn-leaves186269.jpg"
img_array = img_to_array(image_path)
```
